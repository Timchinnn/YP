# Изучение данных

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('https://code.s3.yandex.net/datasets/query_1.csv')
```

```python
df2 = pd.read_csv('https://code.s3.yandex.net/datasets/query_3.csv')
```

# Графики 

```python
a = df.sort_values(by='flights_amount')

palette = ['Airbus A319-100', 'Airbus A321-200', 'Boeing 737-300', 'Boeing 737-300', 'Boeing 777-300', 'Bombardier CRJ-200', 'Cessna 208 Caravan', 'Sukhoi SuperJet-100']

sns.set_style("whitegrid")
plt.figure(figsize=(15, 7))
plt.title('Flights_By_Model', fontsize=20)
sns.barplot(x='model', y='flights_amount', data=a, label="line")
sns.despine(left=True)
plt.xlabel('')
plt.ylabel('flights amount', fontsize=14)
plt.xticks(rotation=30)
plt.show()
```

Cessna 208 Caravan - небольшие самолеты, ,базовая модель перевозит девять пассажиров (до 13 пассажиров) летают по небольшому количеству маршрутво и на короткие расстояния,из-за этого такой спрос;
Bombardier CRJ-200 - самолет на 50 мест, география намного более объемная, чем у "Cessna";
Sukhoi SuperJet-100 - вместимость - 98 мест. 

Как мы видим выделяются топ 3 модели,у которых больше всего полётов 

```python
b = df2.sort_values(by='average_flights', ascending=False)

plt.figure(figsize=(15, 25))
plt.title('Flights By Model', fontsize=20)
sns.barplot(x='average_flights', y='city', data=b, palette='cool')
sns.despine(left=True)
plt.ylabel('')
plt.xlabel('flights amount', fontsize=14)
plt.show()
```

Выберем топ10 городов и построим по ним график. 

```python
top10_cities = df2.sort_values(by='average_flights', ascending=False).head(10)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(15, 7))
plt.title('Top 10 Cities', fontsize=20)
sns.barplot(x='city', y='average_flights', data=top10_cities.sort_values(by='average_flights', ascending=True))
sns.despine(left=True)
plt.xlabel('')
plt.ylabel('flights amount', fontsize=14)
plt.show()
```

Ожидаемо, что Москва впереди — это транспортный пул в любое время года. Питер и Новосибирск занимают 2 и 3 место. Остальные города отличаются слабо.
